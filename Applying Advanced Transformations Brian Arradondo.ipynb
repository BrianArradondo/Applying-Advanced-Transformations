```python
df1[["Hero", "Publisher"]] = df1["Hero|Publisher"].str.split('|',expand=True)
df1.head()
```

```python
df1 = df1.drop(columns = ["Hero|Publisher"]) # Correct and split this column
df1.head() 
```

```python
df1 = df1[["Publisher","Hero","Gender","Race","Alignment","Hair color","Eye color","Skin color", "Measurements"]]
df1.head() # create an order that makes more sense
```

```python
# find and correct the issues with measurements column
df1["Measurements"] = df1["Measurements"].str.replace('"',' ')
df1["Measurements"] = df1["Measurements"].str.replace("{"," ")
df1["Measurements"] = df1["Measurements"].str.replace("}"," ")
df1.head()
```
